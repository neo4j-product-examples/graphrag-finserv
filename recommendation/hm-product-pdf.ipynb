```python
import pandas as pd

product_df = pd.read_csv('https://storage.googleapis.com/neo4j-workshop-data/genai-hm/product.csv')
product_df
```

```python
p_df = product_df[product_df.detailDesc.notnull()][:100]
p_df
```

```python
doc = '\n\n'.join(('##Product' +
                   '\nCode: ' + p_df.productCode.astype(str) +
                   '\nName: ' + p_df.prodName +
                   '\nType: ' + p_df.productTypeName + 
                   '\nGroup: ' + p_df.productGroupName + 
                   '\nGarment Type: ' + p_df.garmentGroupName + 
                   '\nDescription: ' + p_df.detailDesc).tolist())
print(doc[:100000])
```

```python
from fpdf import FPDF

pdf = FPDF()
pdf.add_page()
pdf.set_font("Arial", size = 11)
pdf.write(5, doc.encode('latin-1', 'replace').decode('latin-1'))
pdf.output('hm-products.pdf')
```

```python
import os

output_dir = 'data/product-docs'
if not os.path.exists(output_dir):
    os.makedirs(output_dir)
```

```python
from tqdm import tqdm

for i, row in tqdm(p_df.iterrows()):
    doc = ('Name: ' + row.prodName +
    '\nType: ' + row.productTypeName +
    '\nGroup: ' + row.productGroupName +
    '\nGarment Type: ' + row.garmentGroupName +
    '\nDescription: ' + row.detailDesc)
    #with open(os.path.join(output_dir, f'p{row.productCode}.txt'), 'w') as f:
    #    f.write(doc)
    pdf = FPDF()
    pdf.add_page()
    pdf.set_font("Arial", size = 11)
    pdf.write(5, doc.encode('latin-1', 'replace').decode('latin-1'))
    pdf.output(os.path.join(output_dir, f'p{row.productCode}.pdf'))
```
